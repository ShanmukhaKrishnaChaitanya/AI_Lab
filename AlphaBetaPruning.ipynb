```python
tree = {
    'A': ['B1', 'B2', 'B3'],
    'B1': ['C1', 'C2', 'C3'],
    'B2': ['C4', 'C5', 'C6'],
    'B3': ['C7', 'C8', 'C9']
}

values = {
    'C1': 12, 'C2': 10, 'C3': 3,
    'C4': 5, 'C5': 8, 'C6': 10,
    'C7': 11, 'C8': 2, 'C9': 12
}

def alpha_beta(node, depth, alpha, beta, is_maximizing, values, path, pruned_nodes):
    if depth == 2:
        return values[node], [node]
    
    if is_maximizing:
        best_value = float('-inf')
        best_path = []
        for child in tree[node]:
            value, child_path = alpha_beta(child, depth + 1, alpha, beta, False, values, path, pruned_nodes)
            if value > best_value:
                best_value = value
                best_path = [node] + child_path
            alpha = max(alpha, best_value)
            if beta <= alpha:
                pruned_nodes.append(child)
                break
        return best_value, best_path
    else:
        best_value = float('inf')
        best_path = []
        for child in tree[node]:
            value, child_path = alpha_beta(child, depth + 1, alpha, beta, True, values, path, pruned_nodes)
            if value < best_value:
                best_value = value
                best_path = [node] + child_path
            beta = min(beta, best_value)
            if beta <= alpha:
                pruned_nodes.append(child)
                break
        return best_value, best_path

pruned_nodes = []
optimal_value, optimal_path = alpha_beta('A', 0, float('-inf'), float('inf'), True, values, [], pruned_nodes)
print("Optimal Path:", ' -> '.join(optimal_path))
print("Optimal Value:", optimal_value)
print("Pruned Nodes:", ', '.join(pruned_nodes) if pruned_nodes else "None")
```
